#### DataFrame

Vamos a construir dos instancias de Series:

- `superficie_s`, a partir de un diccionario. Retomamos el ejemplo anterior.

- `poblaci贸n_s`, a partir de dos listas

Luego construimos el DataFrame `provincia_df` a partir de las dos Series.

````python
# Diccionario en python formado por datos de superficies de algunas provincias de Argentina.
# Por cada nombre de provincia un valor de superficie [km虏] asociado.
superficie_dict = {'Buenos Aires': 305907.4, 'C贸rdoba': 164707.8, 'Salta': 155340.5,'Tucum谩n': 22592.1,'La Rioja': 91493.7}

# Serie desde dict y le agregamos el nombre
superficie_s = pd.Series(superficie_dict, name = 'superficie')

# Serie
superficie_s
````

````python
# Diccionario en python formado por datos de provincias de algunas provincias de Argentina.
# Por cada nombre de provincia un valor de poblaci贸n [km虏] asociado.

# Crear lista 1
provincia_lista = ['Buenos Aires','Salta','C贸rdoba', 'Tucum谩n', 'La Rioja']

# Crear lista 2
provincia_pop = [17541141, 1424397, 3760450, 1694656, 393531]

# Serie de provincias desde dos listas
poblacion_s = pd.Series(provincia_pop, index= provincia_lista, name= 'poblacion')

# Mostrar Serie con el nombre poblacion
poblacion_s
````

````python
# Mirar atributos de Series
print('Colecci贸n de valores:',poblacion_s.values)
print('Colecci贸n de etiquetas (indice):',poblacion_s.index)
````

##### Agregar

La funci贸n `Dataframe.aggregate()` se utiliza para aplicar funciones de agregaci贸n en una o m谩s columnas. Algunas de las funciones m谩s utilizadas son las siguientes:
* sum: devuelve la suma de los valores del eje solicitado
* min: devuelve el m铆nimo de los valores del eje solicitado
* max: devuelve el m谩ximo de los valores del eje solicitado
Podemos aplicar la agregaci贸n en un DataFrame, df, como `df.aggregate()` o `df.agg()`.

Estas funciones de agregaci贸n solo funcionan con columnas de tipo num茅rico, pero tambi茅n se podr铆an crear funciones de agregaci贸n para trabajar con otros tipos de datos. Si se desea aplicar agregaci贸n:


```

```

````python
# Indicar el nombre de la funci贸n.
# Por defecto axis = 0 (eje de filas).
# Es decir que se agrega por filas, la funci贸n se aplica a los valores de las filas asociados a cada columna.
numerico_df.agg('max')
````

````python
# Indicar la lista de funciones. Por defecto axis = 0 (eje de filas)
numerico_df.agg(['sum','min','max'])
````

##### Agrupar
Agrupar datos con pandas usando una clave es una operaci贸n com煤n para el an谩lisis de datos.
Para agrupar datos usando una clave, usamos el m茅todo `groupby()` de un DataFrame y especificamos la columna que queremos usar como clave para agrupar:

```

```
Despu茅s de agrupar, normalmente se desea realizar alguna agregaci贸n. Las funciones de agregaci贸n comunes incluyen las siguientes:
* agrupado.sum(): calcula la suma de todas las columnas num茅ricas
* agrupado.mean(): calcula el promedio (media aritm茅tica)
* agrupado.count(): cuenta la cantidad de valores no nulos
* agrupado.agg(['sum', 'mean', 'count']): aplica varias funciones de agregaci贸n a la vez: suma, media y recuento

##### Fusi贸n de DataFrames
La funci贸n `merge()` admite m煤ltiples opciones de uni贸n similares a las operaciones de estilo de base de datos. Fusiona objetos `DataFrame` o `Serie` con nombre con una uni贸n de estilo base de datos.

https://pandas.pydata.org/docs/reference/api/pandas.merge.html#pandas.merge


Estos son los siguientes tipos de uniones:
* La uni贸n interna toma la intersecci贸n de dos o m谩s `DataFrame` o `Serie`. Corresponde a la UNIN INTERNA en el lenguaje de consulta estructurado (SQL).
* La uni贸n externa toma la uni贸n de dos o m谩s `DataFrame` o `Serie`. Corresponde a la UNIN EXTERNA COMPLETA en SQL.
* La uni贸n izquierda usa solo las claves del `DataFrame` o `Serie` izquierdo. Corresponde a la UNIN EXTERNA IZQUIERDA en SQL.
* La uni贸n derecha usa solo las claves del `DataFrame` o `Serie` derecho. Corresponde a la UNIN EXTERNA DERECHA en SQL.


````python

import numpy as np
np.random.seed(1)

# Crear ID en orden
ids = np.arange(1,rev_vino_df.shape[0]+1)

# Crear DF
rev_vino_df['ID'] = ids
df1 = rev_vino_df[['ID','country','price']].head(5)
# Mostrar
display('DF1',df1)
print("--------------------------------")
df2 = rev_vino_df[['ID','points','variety']].head(15)
# Mostrar
display('DF2',df2)
````

````python
datos_inner_df = pd.merge(df1, df2, on='ID', how='inner')
datos_inner_df
````

Como vemos en el fragmento de c贸digo anterior, la funci贸n `pd.merge()` se utiliza para fusionar los dos DataFrames. El argumento `on='ID'` especifica que la columna ID debe utilizarse como clave para unir los DataFrames. El argumento `how='inner'`especifica que debe realizarse una uni贸n interna. Este tipo de uni贸n devuelve solo las filas que tienen valores coincidentes en ambos DataFrames, que, en este caso, son las filas en las que ID coincide tanto en df1 como en df2.

````python
datos_outer_df = pd.merge(df1, df2, on='ID', how='outer')
datos_outer_df
````

Como vemos en el fragmento de c贸digo anterior, el argumento `on='ID'` especifica que la columna ID debe utilizarse como clave en la que fusionar los `DataFrames`. El argumento `how='outer'` especifica que debe realizarse una uni贸n externa completa. Este tipo de uni贸n devuelve todas las filas de ambos `DataFrames`, completando NaN donde no hay coincidencia.

````python
datos_right_df = pd.merge(df1, df2, on='ID', how='right')
datos_right_df
````

El argumento `how='right'` especifica que se debe realizar una uni贸n externa derecha. Este tipo de uni贸n devuelve todas las filas del DataFrame derecho (df1) y las filas coincidentes del DataFrame izquierdo (df2). Si no hay coincidencia, el resultado tendr谩 NaN en las columnas del DataFrame izquierdo.

````python
datos_left_df = pd.merge(df1, df2, on='ID', how='left')
datos_left_df
````

El argumento `how='left'` especifica que se debe realizar una uni贸n externa izquierda. Este tipo de uni贸n devuelve todas las filas del DataFrame izquierdo (df2) y las filas coincidentes del DataFrame derecho (df1). Si no hay coincidencias, el resultado tendr谩 NaN en las columnas del DataFrame derecho.

#### Nota

Manejo de duplicados al fusionar conjuntos de datos
El manejo de claves duplicadas antes de realizar operaciones de fusi贸n es fundamental porque los duplicados pueden generar resultados inesperados, como productos cartesianos, donde las filas se multiplican por la cantidad de entradas coincidentes. ### Tipos de Variables

En estad铆stica las variables, las que se pueden variar, se clasifican en diferentes tipos seg煤n la naturaleza de los datos que representan. Aqu铆 un resumen con ejemplos:

**Variables Cualitativas (o Categ贸ricas)**

Son aquellas que representan caracter铆sticas o categor铆as que no pueden ser medidas con n煤meros. Se dividen en dos tipos:

* Nominales: No tienen un orden espec铆fico.
  * Ejemplo: Color de ojos (azul, verde, marr贸n), estado civil (soltero, casado, divorciado).
* Ordinales: Tienen un orden o jerarqu铆a, pero las diferencias entre valores no son necesariamente iguales.
  * Ejemplo: Nivel de satisfacci贸n (bajo, medio, alto), nivel educativo (primaria, secundaria, universidad).

**Variables Cuantitativas (o Num茅ricas)**

Son aquellas que representan cantidades, medidas con n煤mero y permiten realizar operaciones matem谩ticas. Se dividen en dos tipos:

* Discretas: Toman un valor finito de valores. Son valores enteros contables (no pueden tomar valores intermedios).
  * Ejemplo: N煤mero de integrantes en el hogar (0, 1, 2, 3), cantidad de autos en un estacionamiento.
* Continuas: Pueden tomar cualquier valor dentro de un rango n煤merico, incluyendo decimales.
  * Ejemplo: Altura de una persona (1.75 m), temperatura en grados Celsius (23.5掳C).


En visualizaci贸n, el tipo de variable determina la gr谩fica adecuada para su representaci贸n.

\begin{equation}

Por ejemplo, para la muestra 8, 5 y -1, su media es:

# Medidas de Dispersi贸n o Variabilidad

### Varianza

---

La varianza es un valor num茅rico utilizado para describir cu谩nto var铆an los n煤meros de una distribuci贸n respecto a su media.

Si tenemos un conjunto de valores de una variable, la varianza se calcula de la siguiente forma:

\begin{equation}
\sigma_{n}^{2} = \frac{1}{n} \sum _{i=1}^{n} (x_i - \bar{x})^{2}
\end{equation}

Siendo:

$x_{i}$: cada dato

$\bar{x}$: media de los datos

$n$: n煤mero de datos

Esto es el promedio de **la diferencia elevada al cuadrado entre cada valor y la media**

````python
# Recordar la variable data = np.array([1, 2, 3, 4, 5, 6, 7, 7, 9, 10])

# Calcular varianza con numpy
varianza = np.var(data)

# Mostrar resultados
print("Varianza con numpy:",varianza)

# Calcular varianza con pandas
data_df = pd.DataFrame(data)
varianza = data_df[0].var(ddof=0)

# Mostrar resultados
print("Varianza con pandas:",varianza)
````

### Desv铆o Est谩ndar

---

Es la ra铆z cuadrada de la varianza

\begin{equation}
\sigma_{n} = \sqrt{\frac{1}{n} \sum _{i=1}^{n} (x_i - \bar{x})^{2}}
\end{equation}

El desv铆o est谩ndar es la ra铆z cuadrada de la varianza

El desv铆o es una medida de la dispersi贸n de los datos

**NO** es la desviaci贸n promedio con respecto de la media. Como los desv铆os est谩n elevados al cuadrado los desv铆os muy grandes cuentan m谩s que proporcionalmente.


````python
# Recordar la variable data = np.array([1, 2, 3, 4, 5, 6, 7, 7, 9, 10])

# Calcular std con numpy
std = np.std(data)

# Mostrar resultados
print("Desv铆o Est谩ndar con numpy:",std)

# Calcular std con pandas
data_df = pd.DataFrame(data)
std = data_df[0].std(ddof=0)

# Mostrar resultados
print("Desv铆o Est谩ndar con pandas:",std)
````

## 1.5.1 Codificaci贸n de variables categ贸ricas

En Python para realizar la Codificaci贸n por Etiquetas se puede usar el m贸dulo `LabelEncoder` de la biblioteca `sklearn.preprocessing`. En donde cada categor铆a es un entero que comienza en 0 y termina en n-1, donde n es el n煤mero de categor铆as. Veamos esto con el ejemplo.



Empecemos con `LabelEncoder`

````python
# Importar biblioteca
from sklearn.preprocessing import LabelEncoder

# Inicializar la clase LabelEncoder. Crear instancia
codif_etiqueta = LabelEncoder()

# Entrenar y transformar. LLamar al m茅todo fit_transform
educ_df['LABEL_grado_academico_codificada'] = codif_etiqueta.fit_transform(educ_df['grado_academico'])

# Mostrar datos. Se asigna el n煤mero por orden alfab茅tico.
# En donde cada categor铆a es un entero que comienza en 0 y termina en n-1, donde n es el n煤mero de categor铆as
display(educ_df)
````

2. **Preservaci贸n del orden**: La Codificaci贸n Ordinal conserva la informaci贸n de orden o jerarqu铆a entre las categor铆as. Esto significa que las categor铆as con valores num茅ricos m谩s altos representan un mayor orden o jerarqu铆a en comparaci贸n con las categor铆as con valores num茅ricos m谩s bajos.

3. **Aplicaci贸n en variables categ贸ricas ordinales**: La Codificaci贸n Ordinal es especialmente 煤til cuando se trabaja con variables categ贸ricas que tienen un orden l贸gico predefinido como el ejemplo de grados acad茅micos, clasificaciones (bajo, medio, alto) o tama帽os (peque帽o, mediano, grande), etc.



### Codificaci贸n One-Hot (One-Hot Enconding)

La codificaci贸n One-Hot es una t茅cnica utilizada para convertir variables categ贸ricas en una representaci贸n num茅rica binaria.
````python
# Importar biblioteca
from sklearn.preprocessing import OneHotEncoder

# Inicializar clase OneHotEncoder
codif_onehot = OneHotEncoder()

# Entrenar y transformar los datos.
columnas_codif  = codif_onehot.fit_transform(educ_df[['grado_academico']])


print("Matriz de columnas:",columnas_codif.toarray())

# Crear dataframes con las nuevas caracter铆sticas codificadas One-Hot, y concatenar el DataFrame codificado One-Hot con el DataFrame original:
onehot_df = pd.DataFrame(columnas_codif.toarray(), columns= codif_onehot.get_feature_names_out(['grado_academico']))
educ_onehot_df = pd.concat([educ_df, onehot_df], axis=1)

# Mostrar.
# Aumenta la cantidad de atributos o variables en el conjunto de datos
display(educ_onehot_df)
````

### Estandarizaci贸n

### Normalizaci贸n

La normalizaci贸n es otro proceso com煤n en el preprocesamiento de datos que se utiliza para escalar los valores de una variable. A diferencia de la estandarizaci贸n, que centra los datos alrededor de cero y los escala utilizando la desviaci贸n est谩ndar, la normalizaci贸n ajusta los valores en un rango espec铆fico, generalmente entre 0 y 1.




#### Escalamiento m铆nimo-m谩ximo, (Min-Max Scaling)

Existen varios m茅todos de normalizaci贸n, pero uno de los m谩s comunes es el Min-Max Scaling. En este enfoque, se transforman los datos a un rango espec铆fico utilizando la siguiente f贸rmula:



$X_{nueva} = \frac{X - X_{min}}{X_{max} - X_{min}}$




donde $X$ es el valor observado, $X_{min}$ es el valor m铆nimo en el conjunto de datos y $X_{max}$ es el valor m谩ximo en el conjunto de datos. La f贸rmula normaliza los valores a un rango de [0, 1].

Python ofrece la clase `MinMaxScaler()` de `sklearn.preprocessing`.Veamos como se aplica en el ejemplo edad:


````python
# Importar biblioteca
from sklearn.preprocessing import MinMaxScaler

# Inicializar la clase MinMaxScaler
escalado_MinMax = MinMaxScaler()

# Entrenar y transformar
edad_df['MINMAX_edad'] = escalado_MinMax.fit_transform(edad_df[['edad']])

# Imprimir las primeras 5 instancias
edad_df.head(5)
````

### 1.5.3 Manejo de valores at铆picos (outliers)

La detecci贸n y eliminaci贸n de valores at铆picos es un proceso importante en el preprocesamiento de datos. Definir qu茅 es un valor at铆pico es una tarea dif铆cil. Un definici贸n de valores at铆picos es que son valores que se desv铆an significativamente de la mayor铆a de los puntos de datos en un conjunto de datos.

Estos valores at铆picos pueden ser causados por errores de medici贸n, anomal铆as en los datos o eventos raros. La definici贸n de valores at铆picos implica reconocer los puntos de datos que no se ajustan al comportamiento esperado de los datos y comprender el contexto en el que ocurren.

Los temas a explorar:

En un boxplot, los l铆mites del bigote representan los valores m铆nimo y m谩ximo dentro de un rango aceptable de valores. Estos l铆mites se utilizan para identificar los valores at铆picos (outliers) en el conjunto de datos.  
Cualquier valor que est茅 por encima del l铆mite superior del bigote o por debajo del l铆mite inferior del bigote se considera un valor at铆pico y se muestra individualmente en el diagrama de caja como un punto o asterisco.

Ahora que hemos visto este gr谩fico, estamos empezando a formular algunas hip贸tesis sobre la existencia de valores at铆picos espec铆ficamente en edad. El siguiente paso es utilizar algunos m茅todos estad铆sticos para validar estas
hip贸tesis, comenzando con el m茅todo de IQR, que es esta presente en el Diagrama de caja ya que determina los limites del bigote.

#### M茅todo por Rango Intercuartil (Interquartile Range, IQR)

El IQR es una medida de dispersi贸n estad铆stica que representa el rango entre el tercer cuartil (Q3) y el primer cuartil (Q1) en un conjunto de datos.

Espec铆ficamente, este m茅todo se aplica calculando la siguiente f贸rmula:

* L铆mite inferior del bigote: $Q1 - 1.5 \cdot IQR$
* L铆mite superior del bigote: $Q3 + 1.5  IQR$

donde $IQR$ = rango intercuartil

Recordatorio: c贸mo calcular el IQR

* Calcular Q1 (percentil 25): Identifica el valor por debajo del cual se encuentra el 25 % de los datos ordenados.
* Calcular Q3 (percentil 75): Identifica el valor por debajo del cual se encuentra el 75 % de los datos ordenados.
* Calcular el IQR: Q3 - Q1.



Es importante se帽alar que la elecci贸n del multiplicador (en este caso, 1.5) es algo arbitrario, pero se ha adoptado ampliamente en la pr谩ctica. Ajustar este multiplicador puede hacer que el m茅todo sea m谩s o menos sensible a posibles valores at铆picos. Por ejemplo, el uso de un multiplicador m谩s grande dar铆a como resultado l铆mites m谩s amplios, lo que podr铆a identificar m谩s puntos de datos como posibles valores at铆picos, mientras que un multiplicador m谩s peque帽o har铆a que el m茅todo fuera menos sensible.

````python
# Calcular cuart铆les
Q1 = edad_df['edad'].quantile(0.25)
Q3 = edad_df['edad'].quantile(0.75)

# Calcular rango intercualt铆lico
IQR = Q3 - Q1

# Calcular l铆mites
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Detectar valores at铆picos
atipicos_IQR_df = edad_df[(edad_df['edad'] < limite_inferior) | (edad_df['edad'] > limite_superior)]

# Imprimir resultados
print("Atipicos en 'edad' por IQR:")
atipicos_IQR_df['edad']
````

El m茅todo IQR es simple de aplicar, es conveniente en datos no distribuidos normalmente Sin embargo, tiene ciertas desventajas como el tama帽o de la muestra, especialmente en conjuntos de datos m谩s peque帽os, puede afectar a su precisi贸n a la hora de reflejar la verdadera dispersi贸n de los datos. Detecta at铆picos solo en una variable a la vez, lo que puede ser insuficiente si los valores at铆picos est谩n en combinaciones de variables (at铆picos multivariados).

#### M茅todo de puntuaci贸n Z (Z-score)
El m茅todo de puntuaci贸n Z es una t茅cnica estad铆stica que se utiliza para identificar valores at铆picos univariados en un conjunto de datos midiendo en qu茅 medida los puntos de datos individuales se desv铆an de la media en t茅rminos de desviaciones est谩ndar. Recordemos la f贸rmula de puntuaci贸n Z ya vista:


$$Z = \frac{X-\mu}{\sigma}$$

donde $X$ es el valor observado $\mu$ es la media en el conjunto de datos y $\sigma$ es desviaci贸n est谩ndar en el conjunto de datos.

Normalmente, se elige una puntuaci贸n Z umbral para determinar los valores at铆picos. **Los umbrales que se utilizan habitualmente son Z > 3 o Z < 3, lo que indica que los puntos de datos que se desv铆an m谩s de tres desviaciones est谩ndar de la media se consideran valores at铆picos**.

Gr谩ficas asociadas NORM.png

````python
# Importar biblioteca
from scipy import stats
from sklearn.preprocessing import StandardScaler
import numpy as np

# Calcular puntajes Z (Z-Scores) usando stats . Tambi茅n se podr铆a utilizar StandardScaler() de sklearn
z_puntaje_edad = np.abs(stats.zscore(edad_df['edad'])) ## para trabajar con valores positivos
edad_df['edad_z_puntaje'] = z_puntaje_edad

# Setear umbral de puntaje z
z_umbral = 3

# Identificar valor at铆pico
atipicos_estandar_df = edad_df[edad_df['edad_z_puntaje'] >  z_umbral]

# Imprimir resultados
display("Atipicos en 'edad' por puntaje Z :",atipicos_estandar_df['edad'])
````

Si graficamos las puntuaciones Z para cada observaci贸n y caracter铆stica de los datos, ya podemos empezar a detectar algunos valores at铆picos.
En este gr谩fico de dispersi贸n de puntuaciones Z, cada punto representa la puntuaci贸n Z de un punto de datos individual. La l铆nea discontinua roja indica el umbral de puntuaci贸n Z elegido (en este caso, 3). Los valores at铆picos se identifican como puntos por encima de este umbral.

Continuando con el ejemplo, en edad hay un valor at铆pico claramente capturado.


````python
# Crear gr谩fica
fig, ax = plt.subplots(figsize=(15, 3))
ax.scatter(edad_df['edad_z_puntaje'], edad_df['edad_z_puntaje'], color='blue', label='Z-Scores')
ax.set_title("Puntajes Z (Z-Scores) para 'edad'")
plt.axhline(y=z_umbral, color='red', linestyle='--', label='Umbral')
plt.legend()
````

La puntuaci贸n z es 煤til para detectar valores at铆picos en datos que se distribuyen de forma aproximadamente normal.

**驴C贸mo elegir el umbral correcto para la puntuaci贸n Z?**

Una puntuaci贸n Z le indica a cu谩ntas desviaciones est谩ndar se encuentra un punto de datos con respecto a la media. En una distribuci贸n normal, lo siguiente es cierto:
* Aproximadamente el 68 % de los datos se encuentra dentro de una desviaci贸n est谩ndar de la media.
* Aproximadamente el 95 % de los datos se encuentra dentro de dos desviaciones est谩ndar
* Aproximadamente el 99,7 % de los datos se encuentra dentro de tres desviaciones est谩ndar.

Esto significa que a menudo se utiliza un umbral de puntuaci贸n Z de 3 porque captura valores que est谩n extremadamente alejados
de la media, lo que identifica los valores at铆picos m谩s extremos. En una distribuci贸n perfectamente normal, solo el 0,3 % de los puntos de datos tendr谩 una puntuaci贸n Z mayor que 3 o menor que -3. Esto lo convierte en un umbral razonable para detectar valores at铆picos que es poco probable que formen parte de la distribuci贸n normal de datos.

````python
# M茅todo IQR para identificar valores at铆picos
# Calcular cuartiles
Q1 = certif_cordoba_df['cantidad_certificados'].quantile(0.25)
Q3 = certif_cordoba_df['cantidad_certificados'].quantile(0.75)
# Calcular rango intercualt铆lico
IQR = Q3 - Q1

# Calcular limites
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Detectar valores at铆picos
atipicos_IQR_df = certif_cordoba_df[(certif_cordoba_df['cantidad_certificados'] < limite_inferior) | (certif_cordoba_df['cantidad_certificados'] > limite_superior)]

# Imprimir resultados
print("Atipicos en 'cantidad de certificados' por IQR:")
atipicos_IQR_df
````
