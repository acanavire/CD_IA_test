Idea de gráficos (3).png

* **Cantidades**: El método más común para visualizar cantidades (es decir, valores numéricos que se muestran para un conjunto de categorías) es el uso de barras, ya sea dispuestas vertical u horizontalmente. Sin embargo, en lugar de usar barras.
* **Distribuciones**: Los histogramas proporcionan las visualizaciones más intuitivas de una distribución, pero requiere la elección de parámetros arbitrarios y puede ser engañoso. * **Relaciones x–y**: Los diagramas de dispersión representan la visualización arquetípica cuando queremos mostrar una variable cuantitativa en relación con otra. Cuando el eje x representa el tiempo o una cantidad estrictamente creciente, como una dosis de tratamiento, normalmente dibujamos gráficos de líneas.

#### Gráfico de línea

El gráfico más básico es el gráfico de línea. Los gráficos de líneas se utilizan para mostrar el valor cuantitativo en un intervalo o intervalo de tiempo continuo. Se usa con mayor frecuencia para mostrar tendencias y relaciones (cuando se agrupan con otras líneas).

Para crear este gráfico llamar a la función `plot()` y pasarle los argumentos `x` e `y` que tienen que ser 2 listas o arrays del mismo tamaño.


````python
# Datos
# array - lista de números entre el valor 0 y 5, con un incremento de 0,1 entre los valores
x = np.arange(0,5, 0.1)
print('X:',x)
# Función seno para la lista de números
y = np.sin(x)
print('y:',y)
````

````python
# Para crear gráfica
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y)
plt.show()
````

Además de los datos `x` e `y`, se pueden indicar otros argumentos a la función `plot` para personalizar este tipo de gráfica. Cómo por ejemplo:

````python
# Para crear gráfica personalizada
fig = plt.figure()
ax = plt.axes()
# Cada tipo de grafica puede ser personalizada, mirar su documentación
ax.plot(x, y, color='green', linewidth=5, linestyle='--')
plt.show()
````

También podemos (y siempre deberíamos) agregar títulos a los gráficos y agregar etíquetas(labels) a los ejes para ser lo más claros y explícitos posibles.


````python
# Para crear gráfica personalizada
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y, color='green', linewidth=5, linestyle='--')

# Titulo y etiquetas para los ejes simples
ax.set_title("seno(x) entre 0 y 5")
ax.set_xlabel('etiqueta x')
ax.set_ylabel('etiqueta Y')
plt.show()
````

Pero podriamos también, generar una grilla de ejes simples, tal que en cada celda crear distintos tipos de gráficos para distintos datos. Para ello se usa la función `subplots()`.

**Caso 1**: misma figura, mismos ejes simples, más de un tipo de gráfico.

````python
# Datos
x2 = np.linspace(0, 5)
y2 = x2 / 5
````

````python
# Crear 2 gráficas en misma figura utilizando los mismos ejes
fig, ax = plt.subplots() # equivalente a fig = plt.figure() + ax = plt.axes()

# Crear 1er gráfico de línea en eje simple (1,1)
ax.plot(x, y, color='green', linewidth=5, linestyle='--', label='Seno')
# Crear 2do gráfico de línea en eje simple (1,1)
ax.plot(x2, y2, color='red', linewidth=5, linestyle='-', label='recta')


ax.set_title("seno(x) entre 0 y 5")
ax.set_xlabel('etiqueta x')
ax.set_ylabel('etiqueta Y')

# Para mostrar leyenda
ax.legend()
plt.show()
````

**Caso 2**: misma figura, grilla de ejes simples, más de un tipo de gráfico y distintos datos.

````python
# Crear gráfica con plt.subplots
# Dos variables fig y axes
fig, axes = plt.subplots(nrows=2, ncols=1) # figura y grilla con 2 fias y 1 columna. Ver como una lista.
# Titulo a los subplots
fig.suptitle("Titulo subplots")

# Personalizar cada ejes simples de la grilla
axes[0].plot(x, y, color='green', linewidth=5, linestyle='--', label='Seno')
axes[1].plot(x2, y2, color='red', linewidth=5, linestyle='-', label='recta')

# Para mostrar leyenda
axes[0].legend()
axes[1].legend()
plt.show()
````

#### Histograma

Cuando queremos visualizar la distribución de una variable numérica es muy útil utilizar histogramas. Los histogramas resumen visualmente la distribución de una variable numérica continua midiendo la frecuencia con la que determinados valores aparecen en el conjunto de datos.

En un histograma, el eje x es una línea numérica que se ha dividido en rangos de números o bins. Para cada bin, se dibuja una barra en la que el ancho de la barra representa el rango del bin y la altura de la barra representa el número de puntos de datos incluidos en ese rango.

Para crear está gráfica, llamar la función `hist()`. En cuanto argumentos `x`
puede ser solo un array o una secuencia de arrays que no es necesario que tengan la misma longitud. Veamos el siguiente ejemplo:

````python
# Cantidad de valores
N_valores = 100
# Crear un generador de números aleatorios con una semilla fija para la reproducibilidad
rng = np.random.default_rng(1)
# Distribución normal, media 0 y desvio estándar 1
x = rng.standard_normal(N_valores)
x
````

````python
# Crear Histograma
fig, ax = plt.subplots()
# Titulo
fig.suptitle("Histograma")
ax.hist(x) # defecto bins = 10

# Personalizar más
ax.set_xlabel('bins')
ax.set_ylabel('Frecuencia')
plt.show()
````

¿Qué pasa si usamos un valor muy grande en bins?

````python
# Crear Histograma
fig, ax = plt.subplots()
# Titulo
fig.suptitle("Histograma")
ax.hist(x, bins=200)

# Personalizar más
ax.set_xlabel('bins')
ax.set_ylabel('Frecuencia')
plt.show()
````

¿Y muy chico?

````python
# Crear Histograma
fig, ax = plt.subplots()
# Titulo
fig.suptitle("Histograma")
ax.hist(x, bins=4)


# Personalizar más
ax.set_xlabel('bins')
ax.set_ylabel('Frecuencia')
plt.show()
````

#### Gráfico de Barra
El gráfico de barras clásico utiliza barras horizontales o verticales (gráfico de columnas) para mostrar comparaciones numéricas discretas entre categorías. Un eje del gráfico muestra las categorías específicas que se comparan y el otro eje representa una escala de valores discretos y por lo tanto responden a la pregunta de ¿cuántos? en cada categoría.

Para crear está gráfica, llamar la función `bar()`. Las barras se posicionan en `x` con la alineación dada, barras verticales. Sus dimensiones se dan por altura (`height`) y ancho (`width`).

````python
# Datos
x = ['Manzana', 'Pera', 'Mango', 'Naranja'] # para eje x (categoría)
y = [40, 100, 30, 55] # para eje y (frecuencia)
````

````python
# Crear gráfico de barra
fig, ax = plt.subplots()
fig.suptitle("Gráfico de barra")

ax.bar(x=x,height=y)

ax.set_xlabel('Categorias')
ax.set_ylabel('Frecuencia')
plt.show()
````

#### Diagrama de dispersión

Los diagramas de dispersión usan una colección de puntos dispuestos en coordenadas cartesianas para mostrar valores de dos variables númericas. Al mostrar una variable en cada eje, se puede detectar si existe una relación o correlación entre las dos variables.
Los diagramas de dispersión son ideales cuando se tienen datos numéricos emparejados y se desea ver si una variable afecta a la otra. Sin embargo, recuerde que la correlación no es causal y otra variable inadvertida puede estar influyendo en los resultados.

Para crear está gráfica, llamar la función `scatter()`. En cuanto argumentos `x` ,`y` puede ser un escalar o un array o semejante.



````python
import pandas as pd

# Datos
n = 1000 # Cantidad de valores

# loc=0.0 (media), scale=1.0 (desvio estandar)
x = np.random.normal(size = n)
y = np.random.normal(size = n)
````

````python
# Crear gráfico de barra
fig, ax = plt.subplots()
fig.suptitle("Diagrama de dispersión")

ax.scatter(x, y)
ax.set_xlabel('valores de x')
ax.set_ylabel('valores de y')
plt.show()
````

````python
n = 100
# Datos
x = np.linspace(-1,1,n) + 0.25*np.random.normal(size = n)
y = 3*x + 0.25*np.random.normal(size = n)

# Crear gráfica
fig, ax = plt.subplots()
fig.suptitle("Diagrama de dispersión")

ax.scatter(x, y)
ax.set_xlabel('valores de x')
ax.set_ylabel('valores de y')
plt.show()
````

##### Crear gráfico por una función Axes-level

````python
# Crear gráfico de linea por una función Axes-level
# Por defecto estimator='mean'. Método para agregar múltiples observaciones de la variable 'y' en el mismo nivel 'x'.
# Si se selecciona Ninguno, se extraerán todas las observaciones.
# errorbar=(2 items) ('ci', 95) - Nombre del método errorbar

#Cuando se muestra una estadística de resumen, normalmente es adecuado agregar barras de error, que brindan una pista visual sobre qué tan bien el resumen representa los puntos de datos subyacentes.
sns.lineplot(data=flights, x="year", y="passengers")
````

`matplotlib.axes.Axes.set()`

````python
#  función Axes-level devuelve un objeto matplotlib.axes.Axes
# Recordar los objetos de  matplotlib y los argumentos asociados para el tipo de gráficos
g = sns.lineplot(data=flights_wide)
g.figure.suptitle('Cantidad de pasajeros entre los años 1949 a 1960')
g.figure.set_size_inches((10, 4))
g.set(
    xlabel="Años",
    ylabel="Cantidad de Pasajeros",

)
g
````

**Por las funciones de Matplotlib**

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(10, 4))
fig.suptitle('Cantidad de pasajeros entre los años 1949 a 1960')

sns.lineplot(data=flights_wide, ax=ax)

ax.set_xlabel("Años")
ax.set_ylabel("Cantidad de Pasajeros")
ax.legend(loc= "upper left")

````

Las funciones a nivel de figura ignoran el tamaño de figura predeterminado y parametrizan el tamaño de figura de forma diferente. Al llamar a una función a nivel de figura, puede pasar valores a `height` y `aspect`` para establecer (aproximadamente) el tamaño de cada subgráfico. La ventaja aquí es que el tamaño de la figura se adapta automáticamente cuando agrega variables de facetado. Pero puede resultar confuso.

### Seaborn y Matplotlib

* Matplotlib es una biblioteca proporciona una amplia gama de herramientas para crear visualizaciones altamente personalizables. Seaborn está construida encima de ella. Muchas acciones pueden realizarse solo con las funciones de Seaborn. Otras opciones de personalización pueden requerir el uso directo de Matplotlib.

* Seaborn, proporciona una interfaz más sencilla e intuitiva para crear gráficos estadísticos comunes. Seaborn está diseñado para trabajar con `DataFrame` de Pandas, lo que facilita la creación de visualizaciones con un código mínimo. También ofrece una serie de funciones estadísticas integradas, que permiten a los usuarios realizar fácilmente complejos análisis estadísticos con sus visualizaciones.

* Seaborn, por su parte, ofrece una gama de estilos y paletas de colores predeterminados optimizados para distintos tipos de datos y visualizaciones. Esto facilita a los usuarios la creación de gráficos visualmente atractivos con una personalización mínima.


````python
# Crear histograma con matplotlib
fig, axs = plt.subplots(1,2,figsize=(10, 4), sharey=True)
fig.suptitle("Histograma")
axs[0].hist(tips.loc[tips['sex']=='Male','total_bill'],color='red',alpha=0.7,  bins=4)
axs[1].hist(tips.loc[tips['sex']=='Female','total_bill'],color='blue',alpha=0.7, bins=4)
axs[0].set_xlabel('bins')
axs[0].set_ylabel('Frecuencia')
axs[1].set_xlabel('bins')
plt.show()
````

````python
# # Crear Histograma con seaborn
g = sns.displot(tips, x="total_bill",  bins=4,  palette=['red','blue'], kind='hist', hue='sex',  col='sex')
g.figure.suptitle('Histograma')
g.figure.set_size_inches((10, 4))
g.set( xlabel='bins',ylabel='Frecuencia')
g
````
