## 1.5.2 Métricas

La evaluación correcta de los sistemas es esencial para el éxito del desarrollo. Si algo está mal con la evaluación, el esfuerzo de desarrollo es inútil. ¿Qué puede estar mal?

* Los datos en sí, o cómo los usamos
   * Si se evalúa en datos usados para entrenar o desarrollar
   * Si se evalúa en datos no representativos del caso de uso
   * Si se pierde de vista que hay grupos con distinto comportamiento. Podemos estar ignorando que el sistema es sesgado o injusto
* Las métricas
* Poca o nula significancia estadística (¿ruido o diferencia real?)

Cuando pensamos en las métricas, hay que considerar una serie de cuestiones:
* ¿Qué queremos medir? ¿Qué nos importa en la práctica?
* ¿Nos importan todos los errores por igual o algunos son peores que otros?
* ¿Qué queremos hacer con la salida del sistema? ¿Ordenar, clasificar...?
* ¿Nos importa el resultado global o queremos ver comportamiento por grupos de
muestras? (¿Grupos menos representados, sesgos?)
* ¿Queremos poder interpretar el puntaje que sale del sistema?

* Método alternativo a cross-validation
* Muestreo el set original, con reemplazo
* Para cada set resultante, entreno el modelo, y lo testeo en las muestras no usadas para entrenar
* Calculo el rendimiento como un promedio ponderado sobre las muestras testeadas
