#### Series
---
- Es un objeto similar a un **vector unidimensional**.
- Colección de valores que mantiene un orden. Los valores pueden ser de cualquier tipo de datos, como números enteros, cadenas, objetos de Python, etc.
- Esta formado por una colección de valores y una colección de etiquetas de estos valores denominado **índice** (index).



image.png


La serie más simple se forma a partir de una lista de datos.

Si necesitas reforzar sobre las estructuras de datos en Python, mirar la notebook **Fundamentos** de este curso o click en [Estructura de datos de la documentación oficial de Python](https://docs.python.org/es/3/tutorial/datastructures.html).

````python
# Pasar como argumento la lista = [4, 17, -51, 3]
s = pd.Series([4, 17, -51, 3])
s
````

Como no especificamos un índice para los datos, se crea uno predeterminado que consta de los números enteros del 0 al N - 1 (donde N es la longitud de los datos). Para el ejemplo la serie tiene una colección de N=4 valores y el indice es una colección de enteros de 0 a 3.

Para acceder a la colección de valores y al indice considerar los atributos de la serie `value` e `index` respectivamente:

````python
# Colección de valores
s.values
````

````python
# Colección de etiquetas. Indice
s.index
````

Si tiene datos contenidos en un diccionario de Python, puede crear una serie a partir de ellos pasando el diccionario:

````python
# Diccionario en Python formado por datos de superficies de algunas provincias de Argentina.
# Por cada nombre de provincia un valor de superficie [km²] asociado.
superficie_dict = {'Buenos Aires': 305907.4, 'Córdoba': 164707.8, 'Salta': 155340.5,'Tucumán': 22592.1,'La Rioja': 91493.7}

# Serie desde dict
superficie_s = pd.Series(superficie_dict)

# Serie
superficie_s
````

Cuando solo se pasa un diccionario, el índice de la serie resultante tendrá las claves del diccionario en el orden generado.


````python
# Crear serie desde un diccionario e indice

# Crear lista de provincias
provincias = ['Buenos Aires', 'Córdoba', 'Salta', 'Misiones']

# Crear serie
superficie_s2 = pd.Series(superficie_s, index=provincias)

# Mostrar
superficie_s2
````

En este caso, se colocaron 3 valores encontrados en superficie_s en las ubicaciones correspondientes, pero como no se encontró ningún valor para 'Misiones', aparece como `NaN` (Not a Number), lo que en Pandas se considera para marcar valores faltantes o NA.

Usaremos los términos "faltante" o "NA" para referirnos a los datos faltantes. Las funciones `isnull` y `notnull` en Pandas se usan para detectar datos faltantes:

````python
# Usar la función isnull de Pandas. Se pasa como argumento una serie.
# Se evalúa si el dato en DataFrame es dato faltante.
pd.isnull(superficie_s2)
````

````python
# Usar la función notnull de Pandas. Se pasa como argumento una serie.
# Se evalúa si el dato en DataFrame no es dato faltante.
pd.notnull(superficie_s2)
````

La serie también tiene estos métodos de instancia.

````python
# Función isnull de la serie
superficie_s2.isnull()
````

#### DataFrame
---
- Representa una estructura de datos tabular, similar a una hoja de cálculo, que contiene una colección ordenada de columnas, cada una de las cuales puede ser de un tipo de valor diferente (numérico, cadena, booleano, etc.).
- El `DataFrame` tiene índices de columnas y filas.
- Cada columna de un `DataFrame` es una `Serie`.



image.png

Vamos a construir dos instancias de Series:

- `superficie_s`, a partir de un diccionario. Retomamos el ejemplo anterior.

- `población_s`, a partir de dos listas

Luego construimos el DataFrame `provincia_df` a partir de las dos Series.

````python
# Diccionario en python formado por datos de superficies de algunas provincias de Argentina.
# Por cada nombre de provincia un valor de superficie [km²] asociado.
superficie_dict = {'Buenos Aires': 305907.4, 'Córdoba': 164707.8, 'Salta': 155340.5,'Tucumán': 22592.1,'La Rioja': 91493.7}

# Serie desde dict y le agregamos el nombre
superficie_s = pd.Series(superficie_dict, name = 'superficie')

# Serie
superficie_s
````

````python
# Diccionario en python formado por datos de provincias de algunas provincias de Argentina.
# Por cada nombre de provincia un valor de población [km²] asociado.

# Crear lista 1
provincia_lista = ['Buenos Aires','Salta','Córdoba', 'Tucumán', 'La Rioja']

# Crear lista 2
provincia_pop = [17541141, 1424397, 3760450, 1694656, 393531]

# Serie de provincias desde dos listas
poblacion_s = pd.Series(provincia_pop, index= provincia_lista, name= 'poblacion')

# Mostrar Serie con el nombre poblacion
poblacion_s
````

````python
# Mirar atributos de Series
print('Colección de valores:',poblacion_s.values)
print('Colección de etiquetas (indice):',poblacion_s.index)
````

Para crear un DataFrame desde un diccionario, tener en cuenta que las claves son las etiquetas de las columnas y los valores son los valores de las colecciones asociadas. Para el ejemplo son los valores de las series creadas previamente.

````python
# Crear Dataframe
provincia_df = pd.DataFrame( {'Población':poblacion_s, 'Superficie(km2)':superficie_s}  )
````

````python
# Mostrar
provincia_df
````

Las filas también se pueden recuperar por posición o nombre mediante un par de métodos. Usando las posiciones y el método `iloc` o las etiquetas y el método `loc`.

`iloc` se basa principalmente en la posición de números enteros pero también se puede utilizar con un array booleano.

**Indexación de sola las filas**

````python
# Número entero - escalar
# Fila en la posicion 0
provincia_df.iloc[0]
````

````python
# Lista de números enteros [0]
provincia_df.iloc[[0]]
````

````python
# Lista de números enteros [0,2] , filas de la posición 0 y posición 3

provincia_df.iloc[[0,3]]
````

````python
# Con un objeto de corte (slice)
#  [ini:fin:pasos]
provincia_df.iloc[:3:]
````

````python
# Con una máscara booleana de la misma longitud que el índice. En este caso indice es para las 5 filas
provincia_df.iloc[[True, False, False, False, False]]
````

**Indexación por fila y columna**

Puede combinar los tipos de indexadores para el índice y las columnas.

````python
# Datos de la fila en posición 0 y columna en posición 1
provincia_df.iloc[0 , 1]
````

````python
# Datos de las filas en posición 0 y 2, columna en posición 1
provincia_df.iloc[[0,2] , 1]
````

````python
# Con objeto de corte filas y columnas [ini:fin:pasos , ini:fin:pasos]
provincia_df.iloc[:3, :1]
````

`loc` permite el acceso a un grupo de filas y columnas por etiqueta(s) o un array booleano.

````python
# Ver df
provincia_df
````

````python
# Etiqueta de fila (indice)
provincia_df.loc['Salta']
````

````python
# Lista de etiquetas de filas
provincia_df.loc[['Salta','Buenos Aires']]
````

````python
# Etiqueta de fila y columna [ etiqueta fila (indice) , etiqueta columna ]
provincia_df.loc['Salta' ,'Población']
````

````python
# Por lista de etiquetas
provincia_df.loc[['Salta','Buenos Aires'], ['Población']]
````

````python
# Por objeto de corte
provincia_df.loc[:'Salta':, :'Población']
````

````python
# Lista booleana con la misma longitud que el eje de filas o columnas
provincia_df.loc[[True, False,False,False,False],[True,False]]
````

* La principal diferencia que existe entre loc e iloc es que en loc se usan las etiquetas (los nombres asignados tanto a las filas como a las columnas) mientras que en iloc se usan las posiciones de los elementos (la posición en la fila o la columna, comenzado a contar en 0).

Podemos indexar un DataFrame usando loc y máscaras booleanas

````python
# Filtro por operador relacional. 'Superficie(km2)' > 148008
provincia_df['Superficie(km2)'] > 148008
````

````python
# Por máscara
provincia_df.loc[provincia_df['Superficie(km2)'] > 148008, ['Población', 'Superficie(km2)']]
````

````python
provincia_df.loc[provincia_df['Superficie(km2)'] > 148008, ['Población']]
````

Otros métodos para la selección de valores que devuelve un subconjunto de los datos son `where`, `mask`, y `query`.

````python
# where() para reemplazar valores donde la condición sea Falso.
# provincia_df['Superficie(km2)'] > 148008 si es falso los valores serán 0
provincia_filtro_df1 = provincia_df.where(provincia_df['Superficie(km2)'] > 148008, 0 )

# Mostrar
provincia_filtro_df1
````

````python
# mask reemplaza los valores donde la condición sea Verdadera.
# provincia_df['Superficie(km2)'] > 148008 si es falso los valores serán 0
provincia_filtro_df2 = provincia_df.mask(provincia_df['Superficie(km2)'] > 148008, 0 )

# Mostrar
provincia_filtro_df2
````

````python
# query para consultar las columnas de un DataFrame con una expresión booleana.
provincia_df.query('`Superficie(km2)` > 148008') # es igual a provincia_df[provincia_df['Superficie(km2)']> 148008]
````

`map` aplica una función a un DataFrame elemento por elemento.
Este método aplica una función que acepta y devuelve un escalar a cada elemento de un DataFrame.

Podemos hacerlo utilizando funciones "lambda". Se utilizan para crear funciones anónimas. La expresión

```
lambda argumento(s) : expresion
```
Genera un objeto de función. El objeto sin nombre se comporta como un objeto de función definido con:

```
def <lambda>(parametros):
    return expresion
```
Para más información sobre `lambda`, click en [expresiones lambda en documentación oficial](https://docs.python.org/3/reference/expressions.html#lambda)


````python
# con funciones lambda.
provincia_df.map(lambda x: round(x,2))
````

````python
# También es posible utilizar el mapa con funciones que no son funciones lambda:
provincia_df.map(round, ndigits=1)
````

Los objetos que se pasan a la función son objetos Series cuyo índice es el índice del DataFrame (axis=0) o las columnas del DataFrame (axis=1).

Para más información, click en [apply en documento oficial](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html)

````python
# Numpy
import numpy as np

# Aplicar suma por eje fila, axis = 0.
provincia_df.apply(np.sum, axis=0)
````

#### ¿Cómo resumir un conjunto de datos de un DataFrame?
Un conjunto de datos se puede agregar y agrupar bajo un determinado criterio ya sea para preprocesar datos, analizar, visualizar, entre otros.
Pandas ofrece varias opciones para resumir datos, entre ellos, `agg`, `groupby`, `pivot_table` y `crosstab`.

La agregación y agrupación de datos son técnicas fundamentales en la limpieza y el preprocesamiento de datos, que cumplen varios propósitos críticos. En primer lugar, permiten resumir grandes conjuntos de datos, transformando datos brutos extensos en resúmenes concisos y significativos que facilitan el análisis y la obtención de información.

Estas técnicas también contribuyen a reducir el volumen de datos, mejorar la eficiencia del procesamiento y crear características valiosas para análisis posteriores o modelos de aprendizaje automático.

##### Agrupar
Agrupar datos con pandas usando una clave es una operación común para el análisis de datos.
Para agrupar datos usando una clave, usamos el método `groupby()` de un DataFrame y especificamos la columna que queremos usar como clave para agrupar:

```
agrupado = df.groupby('nombre_columna')
```
Después de agrupar, normalmente se desea realizar alguna agregación. Las funciones de agregación comunes incluyen las siguientes:
* agrupado.sum(): calcula la suma de todas las columnas numéricas
* agrupado.mean(): calcula el promedio (media aritmética)
* agrupado.count(): cuenta la cantidad de valores no nulos
* agrupado.agg(['sum', 'mean', 'count']): aplica varias funciones de agregación a la vez: suma, media y recuento


Veamos ejemplos:

````python
agrupado_por_country = rev_vino_df.groupby('country')
agrupado_por_country
````

Vemos que groupby nos devuelve un objeto Pandas.core.groupby.generic.DataFrameGroupBy.

Sobre este objeto, podemos aplicar directamente funciones de agregación como:

````python
# Aplicar count, para cada grupo de categorias de la columna Country contar los valores no nulos asociados
agrupado_por_country.count().head()
````

````python
# Aplicar mean. Esto funciona para variables númericas
agrupado_por_country.mean().head()
````

````python
# Aplicar mean para variables númericas
agrupado_por_country.mean(numeric_only=True).head()
````

````python
# Para especifica columna
agrupado_por_country['points'].mean().head()
````

También podemos agrupar por múltiples columnas:

````python
# Agrupar por más de 1 columna
agrupado_por_country_prov = rev_vino_df.groupby(['country', 'province'])

# Aplicar count
agrupado_por_country_prov.mean(numeric_only=True).head()
````

Y si no queremos que las variables por las que agrupamos se conviertan en indices y sean una columna más, podemos especificarlo en la función:

````python
# Agrupar por más de 1 columna y modificar el argumento as_index, que por defecto es as_index = True
agrupado_por_country_prov = rev_vino_df.groupby(['country', 'province'], as_index=False)

# Mostrar
agrupado_por_country_prov.mean(numeric_only=True).head()
````

Finalmente, también podemos aplicar distintas funciones de agregación a cada columna.


1. Agrupar el conjunto de datos por pais.

2. Obtener una columna que tenga el precio medio por país y otra que contenga la sumatoria de puntos. (`.mean()` y `.sum()` ).

````python
# Agrupar por pais y agregar de manera determinada para price y points
agrupado_por_country = rev_vino_df.groupby(['country'], as_index=False).agg(
    {'price': ['mean'],
     'points': ['sum']
     })

# Modificar nombre de columna
agrupado_por_country.columns = ["Country",'Price mean', 'Points sum']

# Mostrar
agrupado_por_country.head()
````

Nota Prácticas recomendadas para agrupar
Al agrupar datos en Pandas, hay varias cosas que se deben tener en cuenta para garantizar resultados precisos:
• Datos faltantes: tenga en cuenta los datos faltantes en las columnas utilizadas para agrupar. Pandas excluirá las filas con datos faltantes del resultado agrupado, lo que puede afectar los cálculos finales.
• MultiIndex: al agrupar por varias columnas, Pandas devuelve un índice jerárquico (MultiIndex). Familiarícese con el trabajo con MultiIndex y considere restablecer el índice si es necesario, como lo hemos estado haciendo para simplificar.
• Orden de operaciones: el orden en el que realiza agrupaciones y agregaciones puede afectar los resultados. Tenga en cuenta la secuencia en la que aplica las funciones de agrupación y agregación.
• Agrupación de conjuntos de datos grandes: para conjuntos de datos grandes, la agrupación puede consumir mucha memoria. Considere usar técnicas como fragmentación o procesamiento paralelo para administrar el uso de memoria y el tiempo de cálculo.

````python
print("Tabla Ascendente por Points")
display(agrupado_por_country.sort_values(by=['Points sum'], ascending=[True]).head())
print("")
print("--------------------------------")
print("")
print("Tabla Descendente por Points")
display(agrupado_por_country.sort_values(by=['Points sum'], ascending=[False]).head())
````

###### Tablas dinámicas (Pivot)
La función `pivot_table()` crea una tabla dinámica de estilo hoja de cálculo como `DataFrame`.
Los niveles de la tabla dinámica se almacenarán en objetos MultiIndex (índices jerárquicos) en el índice y las columnas del `DataFrame`. Se requiere un único valor para cada combinación de índice/columna.



```
pd.pivot_table(data, values, index, columns, aggfunc ...)

```
En términos de sintaxis, el expresion anterior representa lo siguiente:
* data :  un conjunto de datos .

##### Fusión de DataFrames
La función `merge()` admite múltiples opciones de unión similares a las operaciones de estilo de base de datos. Fusiona objetos `DataFrame` o `Serie` con nombre con una unión de estilo base de datos.

https://pandas.pydata.org/docs/reference/api/pandas.merge.html#pandas.merge


Estos son los siguientes tipos de uniones:
* La unión interna toma la intersección de dos o más `DataFrame` o `Serie`. Corresponde a la UNIÓN INTERNA en el lenguaje de consulta estructurado (SQL).
* La ​​unión externa toma la unión de dos o más `DataFrame` o `Serie`. Corresponde a la UNIÓN EXTERNA COMPLETA en SQL.
* La unión izquierda usa solo las claves del `DataFrame` o `Serie` izquierdo. Corresponde a la UNIÓN EXTERNA IZQUIERDA en SQL.
* La unión derecha usa solo las claves del `DataFrame` o `Serie` derecho. Corresponde a la UNIÓN EXTERNA DERECHA en SQL.


````python

import numpy as np
np.random.seed(1)

# Crear ID en orden
ids = np.arange(1,rev_vino_df.shape[0]+1)

# Crear DF
rev_vino_df['ID'] = ids
df1 = rev_vino_df[['ID','country','price']].head(5)
# Mostrar
display('DF1',df1)
print("--------------------------------")
df2 = rev_vino_df[['ID','points','variety']].head(15)
# Mostrar
display('DF2',df2)
````

````python
datos_inner_df = pd.merge(df1, df2, on='ID', how='inner')
datos_inner_df
````

Como vemos en el fragmento de código anterior, la función `pd.merge()` se utiliza para fusionar los dos DataFrames. El argumento `on='ID'` especifica que la columna ID debe utilizarse como clave para unir los DataFrames. El argumento `how='inner'`especifica que debe realizarse una unión interna. Este tipo de unión devuelve solo las filas que tienen valores coincidentes en ambos DataFrames, que, en este caso, son las filas en las que ID coincide tanto en df1 como en df2.

````python
datos_outer_df = pd.merge(df1, df2, on='ID', how='outer')
datos_outer_df
````

Como vemos en el fragmento de código anterior, el argumento `on='ID'` especifica que la columna ID debe utilizarse como clave en la que fusionar los `DataFrames`. El argumento `how='outer'` especifica que debe realizarse una unión externa completa. Este tipo de unión devuelve todas las filas de ambos `DataFrames`, completando NaN donde no hay coincidencia.

````python
datos_right_df = pd.merge(df1, df2, on='ID', how='right')
datos_right_df
````

El argumento `how='right'` especifica que se debe realizar una unión externa derecha. Este tipo de unión devuelve todas las filas del DataFrame derecho (df1) y las filas coincidentes del DataFrame izquierdo (df2). Si no hay coincidencia, el resultado tendrá NaN en las columnas del DataFrame izquierdo.

````python
datos_left_df = pd.merge(df1, df2, on='ID', how='left')
datos_left_df
````

El argumento `how='left'` especifica que se debe realizar una unión externa izquierda. Este tipo de unión devuelve todas las filas del DataFrame izquierdo (df2) y las filas coincidentes del DataFrame derecho (df1). Si no hay coincidencias, el resultado tendrá NaN en las columnas del DataFrame derecho.

#### Nota

Manejo de duplicados al fusionar conjuntos de datos
El manejo de claves duplicadas antes de realizar operaciones de fusión es fundamental porque los duplicados pueden generar resultados inesperados, como productos cartesianos, donde las filas se multiplican por la cantidad de entradas coincidentes. Esto no solo puede distorsionar el análisis de datos, sino que también afecta significativamente el rendimiento debido al mayor tamaño del DataFrame resultante.

### Tipos de Variables

En estadística las variables, las que se pueden variar, se clasifican en diferentes tipos según la naturaleza de los datos que representan. Aquí un resumen con ejemplos:

**Variables Cualitativas (o Categóricas)**

Son aquellas que representan características o categorías que no pueden ser medidas con números. Se dividen en dos tipos:

* Nominales: No tienen un orden específico.
  * Ejemplo: Color de ojos (azul, verde, marrón), estado civil (soltero, casado, divorciado).
* Ordinales: Tienen un orden o jerarquía, pero las diferencias entre valores no son necesariamente iguales.
  * Ejemplo: Nivel de satisfacción (bajo, medio, alto), nivel educativo (primaria, secundaria, universidad).

**Variables Cuantitativas (o Numéricas)**

Son aquellas que representan cantidades, medidas con número y permiten realizar operaciones matemáticas. Se dividen en dos tipos:

* Discretas: Toman un valor finito de valores. Son valores enteros contables (no pueden tomar valores intermedios).
  * Ejemplo: Número de integrantes en el hogar (0, 1, 2, 3), cantidad de autos en un estacionamiento.
* Continuas: Pueden tomar cualquier valor dentro de un rango númerico, incluyendo decimales.
  * Ejemplo: Altura de una persona (1.75 m), temperatura en grados Celsius (23.5°C).


En visualización, el tipo de variable determina la gráfica adecuada para su representación.

### Media

----

Dados los n números ${\{x_{1},x_{2},\ldots ,x_{n}\}}$ la media aritmética se define como:

\begin{equation}
    \bar {x} = {\frac {1}{n}} \sum _{i=1}^{n}x_{i} ={\frac {x_{1}+x_{2}+\cdots +x_{n}}{n}}
\end{equation}


Por ejemplo, para la muestra 8, 5 y -1, su media es:

\begin{equation}
    \bar {x} ={\frac {8+5+(-1)}{3} = 4}
\end{equation}

### Mediana

La mediana es el valor del "medio" de una lista  **ordenada** de datos (o el valor que separa la primera mitad y la segunda mitad de una distribución).

Para una lista ordenada la mediana es calculada de diferente manera dependiendo de la cantidad de elementos de la misma:


**Impar**:

[1, 2, 3, 5, **7**, 8, 9, 10, 15]

9 elementos

La mediana es el valor de la posición 5 (la posición del “medio”)

Mediana =  7

**Par**:

[-5, -1, 0, **1**, **2**, 3, 8, 20]

8 elementos

La mediana es la **media de los valores en las dos posiciones centrales**

Mediana =  (1+2)/2 = 1.5

# Medidas de Dispersión o Variabilidad

---




### Varianza

---

La varianza es un valor numérico utilizado para describir cuánto varían los números de una distribución respecto a su media.

Si tenemos un conjunto de valores de una variable, la varianza se calcula de la siguiente forma:

\begin{equation}
\sigma_{n}^{2} = \frac{1}{n} \sum _{i=1}^{n} (x_i - \bar{x})^{2}
\end{equation}

Siendo:

$x_{i}$: cada dato

$\bar{x}$: media de los datos

$n$: número de datos

Esto es el promedio de **la diferencia elevada al cuadrado entre cada valor y la media**

````python
# Recordar la variable data = np.array([1, 2, 3, 4, 5, 6, 7, 7, 9, 10])

# Calcular varianza con numpy
varianza = np.var(data)

# Mostrar resultados
print("Varianza con numpy:",varianza)

# Calcular varianza con pandas
data_df = pd.DataFrame(data)
varianza = data_df[0].var(ddof=0)

# Mostrar resultados
print("Varianza con pandas:",varianza)
````

### Desvío Estándar

---

Es la raíz cuadrada de la varianza

\begin{equation}
\sigma_{n} = \sqrt{\frac{1}{n} \sum _{i=1}^{n} (x_i - \bar{x})^{2}}
\end{equation}

El desvío estándar es la raíz cuadrada de la varianza

El desvío es una medida de la dispersión de los datos

**NO** es la desviación promedio con respecto de la media. Como los desvíos están elevados al cuadrado los desvíos muy grandes cuentan más que proporcionalmente.


````python
# Recordar la variable data = np.array([1, 2, 3, 4, 5, 6, 7, 7, 9, 10])

# Calcular std con numpy
std = np.std(data)

# Mostrar resultados
print("Desvío Estándar con numpy:",std)

# Calcular std con pandas
data_df = pd.DataFrame(data)
std = data_df[0].std(ddof=0)

# Mostrar resultados
print("Desvío Estándar con pandas:",std)
````

## Cuartiles

Los cuartiles dividen un conjunto de datos en cuatro partes iguales, cada una conteniendo el 25% de los datos. Los cuartiles más comúnmente utilizados son:

* Primer cuartil (Q1): También conocido como el cuartil inferior o el percentil 25 (P25), divide los datos de manera que el 25% de los valores sean menores o iguales a este valor.

* Segundo cuartil (Q2): Es el valor que se encuentra en la mediana, es decir, el punto medio de la distribución. También se conoce como el percentil 50 (P50).

* Tercer cuartil (Q3): También conocido como el cuartil superior o el percentil 75 (P75), divide los datos de manera que el 75% de los valores sean menores o iguales a este valor.

Los cuartiles son útiles para comprender la dispersión y la forma de la distribución de los datos.

## 1.5.1 Codificación de variables categóricas

La codificación de variables categóricas es el proceso de transformar datos categóricos (es decir, datos que representan categorías o grupos) en un formato numérico que pueda ser utilizado por los algoritmos de ML. Esto es necesario porque muchos modelos de ML trabajan únicamente con datos numéricos y no pueden interpretar directamente las categorías en texto o etiquetas.

Las variables categóricas son aquellas que toman valores en categorías, como:
* Color de un automóvil (Negro, Blanco, Rojo, Azul, Otros)
* Géneros de libros (Narrativa, Lírica y Dramática)
* Grados académicos (Bachiller, Licenciado, Magíster

Las variables categóricas pueden ser de dos tipos: **nominales y ordinales**. Las nominales no tienen un orden específico, como el  color de un automóvil o géneros de libros. Las ordinales, por otro lado, tienen un orden, como grados académicos en donde el menor grado de estudio es Bachiller y el mayor el Doctorado.

Existen distintas técnicas, aquí exploraremos:

* **Codificación por Etiqueta (Label Encoding)**/**Codificación Ordinal (Ordinal Encoding)**: asigna un número entero a cada categoría de manera arbitraria o en función de su orden o jerarquía.
* **Codificación One-Hot (One-Hot Encoding):** crea una nueva columna binaria para cada categoría y asigna 1 a la categoría presente y 0 a las demás.


Gráficas asociadas.png

Para explorar las tres técnicas utilizaremos el conjunto de datos de grados acádemicos y la libreria scikit-learn de Python.

````python
# Importar biblioteca para la manipulación de datos
import pandas as pd

# Crear el conjunto de datos sobre grados académicos (Bachiller, Licenciado, Magíster y Doctorado).
# Variable categórica ordinal
educ_ls= ["Bachiller","Bachiller","Bachiller", "Licenciado", "Magíster","Magíster","Doctorado"]
data = {'grado_academico': educ_ls}

# Crear dataframe
educ_df = pd.DataFrame(data)

# Mostrar
display(educ_df)
````

### Codificación por Etiqueta (Label encoding)

La Codificación por Etiquetas es una técnica que codifica valores categóricos a números enteros. Técnica simple y rápida de implementar aunque puede introducir relaciones numéricas ficticias en variables categóricas nominales como color de un automóvil (Negro, Blanco, Rojo, Azul, Otros) o géneros de libros (Narrativa, Lírica y Dramática).

En Python para realizar la Codificación por Etiquetas se puede usar el módulo `LabelEncoder` de la biblioteca `sklearn.preprocessing`. En donde cada categoría es un entero que comienza en 0 y termina en n-1, donde n es el número de categorías. Veamos esto con el ejemplo.



Empecemos con `LabelEncoder`

````python
# Importar biblioteca
from sklearn.preprocessing import LabelEncoder

# Inicializar la clase LabelEncoder. Crear instancia
codif_etiqueta = LabelEncoder()

# Entrenar y transformar. LLamar al método fit_transform
educ_df['LABEL_grado_academico_codificada'] = codif_etiqueta.fit_transform(educ_df['grado_academico'])

# Mostrar datos. Se asigna el número por orden alfabético.
# En donde cada categoría es un entero que comienza en 0 y termina en n-1, donde n es el número de categorías
display(educ_df)
````

````python
# Si quisieramos ver las valoras asignados a las categorías. classes_
categorias_codificadas = codif_etiqueta.classes_

# Crear un diccionario para almacenar la correspondencia
correspondencia = {}

# Asignar la correspondencia entre las categorías y los números
for i, categoria in enumerate(categorias_codificadas):
    correspondencia[categoria] = i

# Imprimir la correspondencia
print(correspondencia)
````

La Codificación por Etiquetado no es la mejor técnica. Para el ejemplo, se esperaría desde el grado académico más bajo hasta el más avanzado ('Bachiller'< 'Licenciado'< 'Magíster'<'Doctorado') tener asignado los valores [0,1,2,3] y no [0,2,3,1]. Esto se debe a que asigna por orden alfabético.


Ante este problema, existe la posibilidad de utilizar Codificación Ordinal, útil para variables ordinales.

### Codificación Ordinal (Ordinal Encoding)
Está técnica asignan un valor numérico a cada categoría en función de su orden o jerarquía natural.
En Python, se puede utilizar la clase `OrdinalEncoder` de la biblioteca `sklearn.preprocessing`. Veamos como aplicar al ejemplo de grados académicos:

````python
# Importar biblioteca
from sklearn.preprocessing import OrdinalEncoder

# Inicializar la clase OrdinalEncoder
codif_ordinal = OrdinalEncoder(categories=[['Bachiller','Licenciado','Magíster','Doctorado']], dtype= 'int') # lista ordenada por nivel de menor a mayor

# Entrenar y transformar
educ_df['ORDINAL_grado_academico_codificada']  = codif_ordinal.fit_transform(educ_df[['grado_academico']]) #

# Mostrar
display(educ_df)
````

````python
# Transformando tipo de datos y valores enteros asignados para que comience en 1
educ_df['ORDINAL_grado_academico_codificada'] = educ_df['ORDINAL_grado_academico_codificada'] + 1

# Imprimir los datos originales y los codificados.
# Con Ordinal se logra que el grado académico más bajo sea 'Bachiller'sea 1 y 'Doctorado' sea 4 ('Bachiller'< 'Licenciado'< 'Magíster'<'Doctorado')
display(educ_df)
````

Algunas características clave de la Codificación Ordinal son:

1. **Asignación de valores numéricos**: A cada categoría única en la variable categórica se le asigna un valor numérico único en función de su orden o jerarquía. Los valores numéricos se asignan de manera secuencial, comenzando desde 0, hasta el número total de categorías menos 1.

2. **Preservación del orden**: La Codificación Ordinal conserva la información de orden o jerarquía entre las categorías. Esto significa que las categorías con valores numéricos más altos representan un mayor orden o jerarquía en comparación con las categorías con valores numéricos más bajos.

3. **Aplicación en variables categóricas ordinales**: La Codificación Ordinal es especialmente útil cuando se trabaja con variables categóricas que tienen un orden lógico predefinido como el ejemplo de grados académicos, clasificaciones (bajo, medio, alto) o tamaños (pequeño, mediano, grande), etc.



### Codificación One-Hot (One-Hot Enconding)

La codificación One-Hot es una técnica utilizada para convertir variables categóricas en una representación numérica binaria. Esta técnica se utiliza cuando las variables categóricas no tienen un orden inherente, variables nominales, y se desea representar cada categoría como una columna binaria separada.

La biblioteca `sklearn.preprocessing` en Python proporciona la clase `OneHotEncoder` que se puede utilizar para realizar la codificación One-Hot. A continuación se muestra un ejemplo de cómo utilizar OneHotEncoder:

````python
# Importar biblioteca
from sklearn.preprocessing import OneHotEncoder

# Inicializar clase OneHotEncoder
codif_onehot = OneHotEncoder()

# Entrenar y transformar los datos.
columnas_codif  = codif_onehot.fit_transform(educ_df[['grado_academico']])


print("Matriz de columnas:",columnas_codif.toarray())

# Crear dataframes con las nuevas características codificadas One-Hot, y concatenar el DataFrame codificado One-Hot con el DataFrame original:
onehot_df = pd.DataFrame(columnas_codif.toarray(), columns= codif_onehot.get_feature_names_out(['grado_academico']))
educ_onehot_df = pd.concat([educ_df, onehot_df], axis=1)

# Mostrar.
# Aumenta la cantidad de atributos o variables en el conjunto de datos
display(educ_onehot_df)
````

Está técnica no introduce relaciones ficticias entre categorías aunque aumenta la cantidad de atributos o variables en el conjunto de datos si hay mucha cantidad de categorías asociadas a la variable a codificar.

### Estandarización

La **estandarización**, también conocida como **normalización z-score**, es una técnica comúnmente utilizada en el preprocesamiento de datos para transformar las variables numéricas de manera que tengan una media de 0 y una desviación estándar de 1. su fórmula está determinada por:

$$Z = \frac{X-\mu}{\sigma}$$

donde $X$ es el valor observado $\mu$ es la media en el conjunto de datos y $\sigma$ es desviación estándar en el conjunto de datos.

### Normalización

La normalización es otro proceso común en el preprocesamiento de datos que se utiliza para escalar los valores de una variable. A diferencia de la estandarización, que centra los datos alrededor de cero y los escala utilizando la desviación estándar, la normalización ajusta los valores en un rango específico, generalmente entre 0 y 1.




#### Escalamiento mínimo-máximo, (Min-Max Scaling)

Existen varios métodos de normalización, pero uno de los más comunes es el Min-Max Scaling. En este enfoque, se transforman los datos a un rango específico utilizando la siguiente fórmula:



$X_{nueva} = \frac{X - X_{min}}{X_{max} - X_{min}}$




donde $X$ es el valor observado, $X_{min}$ es el valor mínimo en el conjunto de datos y $X_{max}$ es el valor máximo en el conjunto de datos. La fórmula normaliza los valores a un rango de [0, 1].

Python ofrece la clase `MinMaxScaler()` de `sklearn.preprocessing`.Veamos como se aplica en el ejemplo edad:


````python
# Importar biblioteca
from sklearn.preprocessing import MinMaxScaler

# Inicializar la clase MinMaxScaler
escalado_MinMax = MinMaxScaler()

# Entrenar y transformar
edad_df['MINMAX_edad'] = escalado_MinMax.fit_transform(edad_df[['edad']])

# Imprimir las primeras 5 instancias
edad_df.head(5)
````


Como se puede ver en la fórmula, el escalado mínimo-máximo conserva el orden relativo de los valores, pero los comprime en un rango específico. El escalamiento Min-Max es ideal cuando los datos no siguen una distribución normal y es importante mantener la relación entre los valores originales. Este método ajusta los datos a un rango específico, como [0, 1], siendo útil en modelos sensibles a magnitudes absolutas, métodos basados en distancias (por ejemplo, k-means o K-Nearest Neighbors). Evite usarlo si los datos contienen valores atípicos, ya que estos pueden comprimir el rango de los valores restantes.

### 1.5.3 Manejo de valores atípicos (outliers)

Estos valores atípicos pueden ser causados por errores de medición, anomalías en los datos o eventos raros. La definición de valores atípicos implica reconocer los puntos de datos que no se ajustan al comportamiento esperado de los datos y comprender el contexto en el que ocurren.

#### Método por Rango Intercuartil (Interquartile Range, IQR)

El IQR es una medida de dispersión estadística que representa el rango entre el tercer cuartil (Q3) y el primer cuartil (Q1) en un conjunto de datos.

Específicamente, este método se aplica calculando la siguiente fórmula:

* Límite inferior del bigote: $Q1 - 1.5 \cdot IQR$
* Límite superior del bigote: $Q3 + 1.5 ⋅ IQR$

donde $IQR$ = rango intercuartil

Recordatorio: cómo calcular el IQR

* Calcular Q1 (percentil 25): Identifica el valor por debajo del cual se encuentra el 25 % de los datos ordenados.
* Calcular Q3 (percentil 75): Identifica el valor por debajo del cual se encuentra el 75 % de los datos ordenados.
* Calcular el IQR: Q3 - Q1.



Es importante señalar que la elección del multiplicador (en este caso, 1.5) es algo arbitrario, pero se ha adoptado ampliamente en la práctica. Ajustar este multiplicador puede hacer que el método sea más o menos sensible a posibles valores atípicos. Por ejemplo, el uso de un multiplicador más grande daría como resultado límites más amplios, lo que podría identificar más puntos de datos como posibles valores atípicos, mientras que un multiplicador más pequeño haría que el método fuera menos sensible.

````python
# Calcular cuartíles
Q1 = edad_df['edad'].quantile(0.25)
Q3 = edad_df['edad'].quantile(0.75)

# Calcular rango intercualtílico
IQR = Q3 - Q1

# Calcular límites
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Detectar valores atípicos
atipicos_IQR_df = edad_df[(edad_df['edad'] < limite_inferior) | (edad_df['edad'] > limite_superior)]

# Imprimir resultados
print("Atipicos en 'edad' por IQR:")
atipicos_IQR_df['edad']
````

El método IQR es simple de aplicar, es conveniente en datos no distribuidos normalmente Sin embargo, tiene ciertas desventajas como el tamaño de la muestra, especialmente en conjuntos de datos más pequeños, puede afectar a su precisión a la hora de reflejar la verdadera dispersión de los datos. Detecta atípicos solo en una variable a la vez, lo que puede ser insuficiente si los valores atípicos están en combinaciones de variables (atípicos multivariados).

#### Método de puntuación Z (Z-score)
El método de puntuación Z es una técnica estadística que se utiliza para identificar valores atípicos univariados en un conjunto de datos midiendo en qué medida los puntos de datos individuales se desvían de la media en términos de desviaciones estándar. Recordemos la fórmula de puntuación Z ya vista:


$$Z = \frac{X-\mu}{\sigma}$$

donde $X$ es el valor observado $\mu$ es la media en el conjunto de datos y $\sigma$ es desviación estándar en el conjunto de datos.

Normalmente, se elige una puntuación Z umbral para determinar los valores atípicos. **Los umbrales que se utilizan habitualmente son Z > 3 o Z <− 3, lo que indica que los puntos de datos que se desvían más de tres desviaciones estándar de la media se consideran valores atípicos**.

Gráficas asociadas NORM.png

````python
# Importar biblioteca
from scipy import stats
from sklearn.preprocessing import StandardScaler
import numpy as np

# Calcular puntajes Z (Z-Scores) usando stats . También se podría utilizar StandardScaler() de sklearn
z_puntaje_edad = np.abs(stats.zscore(edad_df['edad'])) ## para trabajar con valores positivos
edad_df['edad_z_puntaje'] = z_puntaje_edad

# Setear umbral de puntaje z
z_umbral = 3

# Identificar valor atípico
atipicos_estandar_df = edad_df[edad_df['edad_z_puntaje'] >  z_umbral]

# Imprimir resultados
display("Atipicos en 'edad' por puntaje Z :",atipicos_estandar_df['edad'])
````

Si graficamos las puntuaciones Z para cada observación y característica de los datos, ya podemos empezar a detectar algunos valores atípicos.
En este gráfico de dispersión de puntuaciones Z, cada punto representa la puntuación Z de un punto de datos individual. La línea discontinua roja indica el umbral de puntuación Z elegido (en este caso, 3). Los valores atípicos se identifican como puntos por encima de este umbral.

Continuando con el ejemplo, en edad hay un valor atípico claramente capturado.


````python
# Crear gráfica
fig, ax = plt.subplots(figsize=(15, 3))
ax.scatter(edad_df['edad_z_puntaje'], edad_df['edad_z_puntaje'], color='blue', label='Z-Scores')
ax.set_title("Puntajes Z (Z-Scores) para 'edad'")
plt.axhline(y=z_umbral, color='red', linestyle='--', label='Umbral')
plt.legend()
````

La puntuación z es útil para detectar valores atípicos en datos que se distribuyen de forma aproximadamente normal.

**¿Cómo elegir el umbral correcto para la puntuación Z?**

Una puntuación Z le indica a cuántas desviaciones estándar se encuentra un punto de datos con respecto a la media. En una distribución normal, lo siguiente es cierto:
* Aproximadamente el 68 % de los datos se encuentra dentro de una desviación estándar de la media.
* Aproximadamente el 95 % de los datos se encuentra dentro de dos desviaciones estándar
* Aproximadamente el 99,7 % de los datos se encuentra dentro de tres desviaciones estándar.

Esto significa que a menudo se utiliza un umbral de puntuación Z de 3 porque captura valores que están extremadamente alejados
de la media, lo que identifica los valores atípicos más extremos. En una distribución perfectamente normal, solo el 0,3 % de los puntos de datos tendrá una puntuación Z mayor que 3 o menor que -3. Esto lo convierte en un umbral razonable para detectar valores atípicos que es poco probable que formen parte de la distribución normal de datos.

### Eliminación de valores atípicos
La eliminación de valores atípicos se refiere al proceso de remover puntos de datos en un conjunto de datos que se consideran inusualmente extremos o desviados del patrón general de los datos. La eliminación de valores atípicos tiene sus pros y sus contras.
Por un lado, es la forma más sencilla de lidiar con valores extremos. Por otro lado, conduce a una reducción en el tamaño de la muestra y a una posible pérdida de información valiosa. Además, si los valores atípicos no son errores genuinos sino que reflejan una variabilidad legítima en los datos, su eliminación puede introducir sesgo.

Volviendo al ejemplo de edad, vamos a eliminar valores atipicos por Método IQR:

````python
# Seleccionar edad dentro de los limites
datos_sinAtipicos_IQR_df = edad_df[(edad_df['edad'] >= limite_inferior) & (edad_df['edad'] <= limite_superior)]
display("Estadística para 'edad' con valores atípicos ",edad_df['edad'].describe())
display("Estadística para 'edad' sin valores atípicos por IQR ",datos_sinAtipicos_IQR_df['edad'].describe())
````

Los cambios observados después de la eliminación de los valores atípicos se describen a continuación:
* **Cambio de la edad media**: la edad media, después de eliminar los valores atípicos, disminuyó ligeramente de 33,75 a
aproximadamente 28,27. Esta reducción sugiere que los valores atípicos eliminados eran individuos mayores.
* **Cambio de la desviación estándar para la edad**: la desviación estándar para la edad disminuyó de 18,90 a 8,75, lo que indica que la dispersión de edades se volvió ligeramente más estrecha después de la eliminación de los valores atípicos, lo que
probablemente contribuyó a una mayor variabilidad en el conjunto de datos original.
*  **Valores mínimos y máximos de edad**: la edad mínima permaneció igual, 18, mientras que la edad máxima disminuyó de 90 a 45, lo que indica que los individuos mayores (posibles valores atípicos) se eliminaron durante el manejo de los valores atípicos.


La eliminación de valores atípicos condujo a una disminución tanto de la edad media como de la desviación estándar, así como a un ligero aumento de la puntuación media de la prueba. Si bien la eliminación de valores atípicos puede mejorar la calidad de los datos, especialmente cuando se deben a errores de entrada de datos o imprecisiones de medición, también reduce la variabilidad del conjunto de datos. Si los valores atípicos representan una variabilidad real en la población, su eliminación podría distorsionar la imagen general de los datos. Para el ejemplo de edad, no estaríamos contemplado a los adultos mayores. Por lo tanto, se debe considerar cuidadosamente si los valores atípicos representan puntos de datos genuinos o errores.




**Impacto de los valores atípicos**

* **Estadísticas de resumen distorsionadas**: los valores atípicos pueden distorsionar significativamente las estadísticas de resumen, dando una impresión errónea de las tendencias centrales de los datos:

  * Media y mediana: la media, una medida común de tendencia central, puede verse muy afectada por los valores atípicos. Un valor atípico con un valor mucho mayor o menor que el resto puede inclinar la media en su dirección. Por otro lado, la mediana está determinada por el valor medio de un conjunto de datos ordenados. Sirve efectivamente como el punto central que divide los datos en dos mitades iguales, haciéndolos menos susceptibles a la influencia de valores extremos.
  * Varianza y desviación estándar: los valores atípicos pueden inflar la varianza y la desviación estándar, haciendo que los datos parezcan más dispersos de lo que realmente están. Esto puede tergiversar la variabilidad de la mayoría de los datos.
* **Visualizaciones engañosas**: los valores atípicos pueden distorsionar la escala y la forma de las visualizaciones, lo que genera una interpretación errónea:
  * Diagramas de caja: los valores atípicos pueden hacer que los diagramas de caja se extiendan excesivamente, lo que hace que la mayor parte de los datos parezcan comprimidos. Esto puede hacer que la distribución parezca menos dispersa de lo que realmente es.
  * Histogramas: los valores atípicos pueden generar la creación de intervalos que capturan solo unos pocos valores extremos, lo que hace que otros intervalos parezcan desproporcionadamente pequeños y que la forma de la distribución esté distorsionada.

* **Influencia en el rendimiento del modelo**: los valores atípicos pueden afectar negativamente el rendimiento de los modelos. por ejemplo modelo de regresión y modelos de clustering.
