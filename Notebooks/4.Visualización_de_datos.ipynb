#### Gr谩fico de l铆nea

El gr谩fico m谩s b谩sico es el gr谩fico de l铆nea. Los gr谩ficos de l铆neas se utilizan para mostrar el valor cuantitativo en un intervalo o intervalo de tiempo continuo. Se usa con mayor frecuencia para mostrar tendencias y relaciones (cuando se agrupan con otras l铆neas).

Para crear este gr谩fico llamar a la funci贸n `plot()` y pasarle los argumentos `x` e `y` que tienen que ser 2 listas o arrays del mismo tama帽o.


````python
# Datos
# array - lista de n煤meros entre el valor 0 y 5, con un incremento de 0,1 entre los valores
x = np.arange(0,5, 0.1)
print('X:',x)
# Funci贸n seno para la lista de n煤meros
y = np.sin(x)
print('y:',y)
````

````python
# Para crear gr谩fica
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y)
plt.show()
````

Adem谩s de los datos `x` e `y`, se pueden indicar otros argumentos a la funci贸n `plot` para personalizar este tipo de gr谩fica. C贸mo por ejemplo:

````python
# Para crear gr谩fica personalizada
fig = plt.figure()
ax = plt.axes()
# Cada tipo de grafica puede ser personalizada, mirar su documentaci贸n
ax.plot(x, y, color='green', linewidth=5, linestyle='--')
plt.show()
````

Tambi茅n podemos (y siempre deber铆amos) agregar t铆tulos a los gr谩ficos y agregar et铆quetas(labels) a los ejes para ser lo m谩s claros y expl铆citos posibles.


````python
# Para crear gr谩fica personalizada
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y, color='green', linewidth=5, linestyle='--')

# Titulo y etiquetas para los ejes simples
ax.set_title("seno(x) entre 0 y 5")
ax.set_xlabel('etiqueta x')
ax.set_ylabel('etiqueta Y')
plt.show()
````

Pero podriamos tambi茅n, generar una grilla de ejes simples, tal que en cada celda crear distintos tipos de gr谩ficos para distintos datos. Para ello se usa la funci贸n `subplots()`.

**Caso 1**: misma figura, mismos ejes simples, m谩s de un tipo de gr谩fico.

````python
# Datos
x2 = np.linspace(0, 5)
y2 = x2 / 5
````

````python
# Crear 2 gr谩ficas en misma figura utilizando los mismos ejes
fig, ax = plt.subplots() # equivalente a fig = plt.figure() + ax = plt.axes()

# Crear 1er gr谩fico de l铆nea en eje simple (1,1)
ax.plot(x, y, color='green', linewidth=5, linestyle='--', label='Seno')
# Crear 2do gr谩fico de l铆nea en eje simple (1,1)
ax.plot(x2, y2, color='red', linewidth=5, linestyle='-', label='recta')


ax.set_title("seno(x) entre 0 y 5")
ax.set_xlabel('etiqueta x')
ax.set_ylabel('etiqueta Y')

# Para mostrar leyenda
ax.legend()
plt.show()
````

**Caso 2**: misma figura, grilla de ejes simples, m谩s de un tipo de gr谩fico y distintos datos.

````python
# Crear gr谩fica con plt.subplots
# Dos variables fig y axes
fig, axes = plt.subplots(nrows=2, ncols=1) # figura y grilla con 2 fias y 1 columna. Ver como una lista.
# Titulo a los subplots
fig.suptitle("Titulo subplots")

# Personalizar cada ejes simples de la grilla
axes[0].plot(x, y, color='green', linewidth=5, linestyle='--', label='Seno')
axes[1].plot(x2, y2, color='red', linewidth=5, linestyle='-', label='recta')

# Para mostrar leyenda
axes[0].legend()
axes[1].legend()
plt.show()
````

#### Histograma

Cuando queremos visualizar la distribuci贸n de una variable num茅rica es muy 煤til utilizar histogramas. Los histogramas resumen visualmente la distribuci贸n de una variable num茅rica continua midiendo la frecuencia con la que determinados valores aparecen en el conjunto de datos.

En un histograma, el eje x es una l铆nea num茅rica que se ha dividido en rangos de n煤meros o bins. Para cada bin, se dibuja una barra en la que el ancho de la barra representa el rango del bin y la altura de la barra representa el n煤mero de puntos de datos incluidos en ese rango.

Para crear est谩 gr谩fica, llamar la funci贸n `hist()`. En cuanto argumentos `x`
puede ser solo un array o una secuencia de arrays que no es necesario que tengan la misma longitud. Veamos el siguiente ejemplo:

````python
# Cantidad de valores
N_valores = 100
# Crear un generador de n煤meros aleatorios con una semilla fija para la reproducibilidad
rng = np.random.default_rng(1)
# Distribuci贸n normal, media 0 y desvio est谩ndar 1
x = rng.standard_normal(N_valores)
x
````

````python
# Crear Histograma
fig, ax = plt.subplots()
# Titulo
fig.suptitle("Histograma")
ax.hist(x) # defecto bins = 10

# Personalizar m谩s
ax.set_xlabel('bins')
ax.set_ylabel('Frecuencia')
plt.show()
````

驴Qu茅 pasa si usamos un valor muy grande en bins?

````python
# Crear Histograma
fig, ax = plt.subplots()
# Titulo
fig.suptitle("Histograma")
ax.hist(x, bins=4)


# Personalizar m谩s
ax.set_xlabel('bins')
ax.set_ylabel('Frecuencia')
plt.show()
````

#### Gr谩fico de Barra
El gr谩fico de barras cl谩sico utiliza barras horizontales o verticales (gr谩fico de columnas) para mostrar comparaciones num茅ricas discretas entre categor铆as. Un eje del gr谩fico muestra las categor铆as espec铆ficas que se comparan y el otro eje representa una escala de valores discretos y por lo tanto responden a la pregunta de 驴cu谩ntos? en cada categor铆a.

Para crear est谩 gr谩fica, llamar la funci贸n `bar()`. Las barras se posicionan en `x` con la alineaci贸n dada, barras verticales. Sus dimensiones se dan por altura (`height`) y ancho (`width`).

````python
# Datos
x = ['Manzana', 'Pera', 'Mango', 'Naranja'] # para eje x (categor铆a)
y = [40, 100, 30, 55] # para eje y (frecuencia)
````

````python
# Crear gr谩fico de barra
fig, ax = plt.subplots()
fig.suptitle("Gr谩fico de barra")

ax.bar(x=x,height=y)

ax.set_xlabel('Categorias')
ax.set_ylabel('Frecuencia')
plt.show()
````

#### Diagrama de dispersi贸n

Los diagramas de dispersi贸n usan una colecci贸n de puntos dispuestos en coordenadas cartesianas para mostrar valores de dos variables n煤mericas. Al mostrar una variable en cada eje, se puede detectar si existe una relaci贸n o correlaci贸n entre las dos variables.
Los diagramas de dispersi贸n son ideales cuando se tienen datos num茅ricos emparejados y se desea ver si una variable afecta a la otra. Sin embargo, recuerde que la correlaci贸n no es causal y otra variable inadvertida puede estar influyendo en los resultados.

Para crear est谩 gr谩fica, llamar la funci贸n `scatter()`. En cuanto argumentos `x` ,`y` puede ser un escalar o un array o semejante.



````python
import pandas as pd

# Datos
n = 1000 # Cantidad de valores

# loc=0.0 (media), scale=1.0 (desvio estandar)
x = np.random.normal(size = n)
y = np.random.normal(size = n)
````

````python
# Crear gr谩fico de barra
fig, ax = plt.subplots()
fig.suptitle("Diagrama de dispersi贸n")

ax.scatter(x, y)
ax.set_xlabel('valores de x')
ax.set_ylabel('valores de y')
plt.show()
````

````python
n = 100
# Datos
x = np.linspace(-1,1,n) + 0.25*np.random.normal(size = n)
y = 3*x + 0.25*np.random.normal(size = n)

# Crear gr谩fica
fig, ax = plt.subplots()
fig.suptitle("Diagrama de dispersi贸n")

ax.scatter(x, y)
ax.set_xlabel('valores de x')
ax.set_ylabel('valores de y')
plt.show()
````

### Seaborn y Matplotlib

* Matplotlib es una biblioteca proporciona una amplia gama de herramientas para crear visualizaciones altamente personalizables. Seaborn est谩 construida encima de ella. Muchas acciones pueden realizarse solo con las funciones de Seaborn. Otras opciones de personalizaci贸n pueden requerir el uso directo de Matplotlib.

* Seaborn, proporciona una interfaz m谩s sencilla e intuitiva para crear gr谩ficos estad铆sticos comunes. Seaborn est谩 dise帽ado para trabajar con `DataFrame` de Pandas, lo que facilita la creaci贸n de visualizaciones con un c贸digo m铆nimo. Tambi茅n ofrece una serie de funciones estad铆sticas integradas, que permiten a los usuarios realizar f谩cilmente complejos an谩lisis estad铆sticos con sus visualizaciones.

* Seaborn, por su parte, ofrece una gama de estilos y paletas de colores predeterminados optimizados para distintos tipos de datos y visualizaciones. Esto facilita a los usuarios la creaci贸n de gr谩ficos visualmente atractivos con una personalizaci贸n m铆nima.


````python
# Crear histograma con matplotlib
fig, axs = plt.subplots(1,2,figsize=(10, 4), sharey=True)
fig.suptitle("Histograma")
axs[0].hist(tips.loc[tips['sex']=='Male','total_bill'],color='red',alpha=0.7,  bins=4)
axs[1].hist(tips.loc[tips['sex']=='Female','total_bill'],color='blue',alpha=0.7, bins=4)
axs[0].set_xlabel('bins')
axs[0].set_ylabel('Frecuencia')
axs[1].set_xlabel('bins')
plt.show()
````

````python
# # Crear Histograma con seaborn
g = sns.displot(tips, x="total_bill",  bins=4,  palette=['red','blue'], kind='hist', hue='sex',  col='sex')
g.figure.suptitle('Histograma')
g.figure.set_size_inches((10, 4))
g.set( xlabel='bins',ylabel='Frecuencia')
g
````
